```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = 'k'
plt.rcParams['axes.labelcolor'] = 'k'
plt.rcParams['text.usetex'] = True

marks = [
    ',',  # pixel
    'o',  # circle
    'v',  # triangle_down
    '^',  # triangle_up
    '<',  # triangle_left
    '>',  # triangle_right
    '1',  # tri_down
    '2',  # tri_up
    '3',  # tri_left
    '4',  # tri_right
    '8',  # octagon
    's',  # square
    'p',  # pentagon
    'P',  # plus (filled)
    '*',  # star
    'h',  # hexagon1
    'H',  # hexagon2
    '+',  # plus
    'x',  # x
    'X',  # x (filled)
    'D',  # diamond
    'd',  # thin_diamond
    '|',  # vline
    '_',  # hline
]
```

```python
graph_qubit = 4
total_qubits = 200
avg_gates_per_layer = int((total_qubits//graph_qubit)*0.5)
layers = 750
num_shots = 2_00

num_graphs = 2

entropies = []


for i in range(num_graphs):
    entropy = np.loadtxt(rf'entropy-q4-layers{layers}-qubits{total_qubits}-shots{num_shots}-graph{i + 1}.txt')
    entropies.append(entropy)
    i += 1
```

```python

fig, ax = plt.subplots(figsize=(4,3))

t = np.arange(0, layers, 1)

# Plot each entropy trace with cubehelix colors
for i in range(len(entropies)):
    j = i+5
    ax.plot(t, entropies[i], label=str(i + 1))


for spine in ax.spines.values():
    spine.set_edgecolor('gray')

plt.tick_params(top=True, left=True, right=True, bottom=True,
               direction="in", axis='both', which='both', 
               labelsize=15, color='k')

font2 = {'family':'Open Sans', 'color':'black', 'size':18}

plt.xlabel(r"$t$", fontdict=font2)
plt.ylabel(r"$\langle S(t) \rangle$", fontdict=font2)

lstyle = {'family':'Open Sans', 'size': 16}

# Modified legend placement
plt.legend(
    fontsize=16,  # Slightly smaller font for external legend
    ncol=1,  # Fewer columns for vertical space
    loc='best', 
    # bbox_to_anchor=(1.02, 0.9),  # Positions legend outside axes
    frameon=False,
    prop = lstyle,
    borderaxespad=0.  # Removes padding between axes and legend
)

# ax.set_ylim(0.5,)
fig.set_dpi(120)
ax.set_yticks([0, 20, 40, 60, 80, 100])
ax.set_yticklabels([0, 20, 40, 60, 80, 100])

plt.tight_layout(rect=[0, 0, 1, 1])  # Adjust right margin to make space
plt.show()
```
