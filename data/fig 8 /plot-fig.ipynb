```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = 'k'
plt.rcParams['axes.labelcolor'] = 'k'
plt.rcParams['text.usetex'] = True

marks = [
    '.',  # point
    ',',  # pixel
    'o',  # circle
    'v',  # triangle_down
    '^',  # triangle_up
    '<',  # triangle_left
    '>',  # triangle_right
    '1',  # tri_down
    '2',  # tri_up
    '3',  # tri_left
    '4',  # tri_right
    '8',  # octagon
    's',  # square
    'p',  # pentagon
    'P',  # plus (filled)
    '*',  # star
    'h',  # hexagon1
    'H',  # hexagon2
    '+',  # plus
    'x',  # x
    'X',  # x (filled)
    'D',  # diamond
    'd',  # thin_diamond
    '|',  # vline
    '_',  # hline
]
```

```python
graph_qubit = 5
total_qubits = 200
layers = 500
num_shots = 2_00

alpha = [0.4, 0.5, 0.6]

avg_gates_per_layer = [int((total_qubits//graph_qubit)*ai) for ai in alpha]

entropy = np.loadtxt('entropy_vary_alpha.txt')
```

```python
fig, ax = plt.subplots(figsize=(4,2.5))

t = np.arange(0, layers, 1)

for i in range(len(alpha)):
    ax.plot(t, entropy[:, i], label = rf'$\alpha = $ {alpha[i]}')

for spine in ax.spines.values():
    spine.set_edgecolor('gray')

plt.tick_params(top=True, left=True, right=True, bottom=True,
               direction="in", axis='both', which='both', 
               labelsize=14, color='k')

font2 = {'family':'serif', 'color':'black', 'size':18}

plt.xlabel(r"$t$", fontdict=font2, usetex='True')
plt.ylabel(r"$\langle S(t)\rangle$", fontdict=font2, usetex='True')

# Modified legend placement
plt.legend(
    fontsize=13,  # Slightly smaller font for external legend
    ncol=1,  # Fewer columns for vertical space
    loc='best', 
    # bbox_to_anchor=(1.02, 0.9),  # Positions legend outside axes
    frameon=False,
    borderaxespad=0.,  # Removes padding between axes and legend
    # handletextpad=0.2
)

fig.set_dpi(120)

plt.tight_layout(rect=[0, 0, 1, 1])  # Adjust right margin to make space
plt.show()
```
