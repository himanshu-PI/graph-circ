```python
graph_qubit = 5
total_qubits = 200
avg_gates_per_layer = int((total_qubits//graph_qubit)*0.5)
layers = 5*int(total_qubits/6)
num_shots = 5_00

graphs = Q5_graph

commutator = []
i = 1
for Gi in graphs:

    graph_state = Gi

    C = average_otoc_evolution(Gi, 
                               total_qubits, 
                               avg_gates_per_layer, 
                               layers, 
                               num_shots) 
    commutator.append(C)
    i+=1


```

```python
pal = sns.color_palette("rocket", as_cmap=True)

fig, axs = plt.subplots(1,4, figsize=(17.5, 7.5))

for i in range(4):

    data = commutator[i].copy()
    vmin = 5e-2
    data[data <= 0] = vmin
    # Plot the boolean matrix
    im = axs[i].imshow(data, cmap= pal, interpolation="none", origin='lower', norm = LogNorm(vmin = vmin, vmax = 1))

    axs[i].grid(which="minor", color="gray", linewidth=0.5)
    axs[i].tick_params(which="both", bottom=False, left=False, labelbottom=False, labelleft=False)


# ----- ADD THIS FOR A SINGLE COLORBAR -----
cbar = fig.colorbar(
    im,
    ax=axs.ravel().tolist(),
    location='right',
    shrink=0.35, anchor = (-0.2, 0.5), aspect = 15  # <-- make colorbar shorter
)

cbar.ax.tick_params(labelsize=12)

plt.tick_params(top=True, left=True, right=True, bottom=True,
               direction="in", axis='both', which='both', 
               labelsize=15, color='k')

font2 = {'family': 'serif', 'color': 'black', 'size': 18}
plt.show()
```
