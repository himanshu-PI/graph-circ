```python
# 4 qubit graph calculation
```

```python
# parameters : 

# Note : system size is reduced 

graph_qubit = 4
total_qubits = 500
sparsity = 0.5
avg_gates_per_layer = int((total_qubits//graph_qubit) * sparsity)
layers = 2*int(total_qubits/6)
num_shots = 500

graphs = Q4_graph


vB = []

i = 1

for Gi in graphs:
    graph_state = Gi
    vb = butterfly_vel(Gi, total_qubits, avg_gates_per_layer, layers, num_shots) 
    vB.append(vb)
    i+=1
```

```python
# cross edge calculation 

lattice  = [i for i in range(graph_qubit)]


graph_conn = []

for Gi in graphs:

    no_conn = 0
    for i in range(1,graph_qubit):

        partA = graph_qubit[:i]
        partB = graph_qubit[i:]
        # print(partA,partB)
        no_conn += count_cross_edges(Gi.edges, partA, partB)
    
    graph_conn.append(no_conn)


print(graph_conn)
```
